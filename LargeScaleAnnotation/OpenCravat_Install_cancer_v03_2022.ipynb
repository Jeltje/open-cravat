## Install OpenCravat and Annotators

### Install open-cravat

```python
pip install open-cravat
```

### Install base modules and reporters

```python
%%bash
oc module install-base
oc module install -y excelreporter
oc module install -y textreporter
oc module install -y tsvreporter
```

### Install Clinvar (as a test for specific modules)

```python
%%bash
echo y | oc module install clinvar
```
